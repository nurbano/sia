```python
#Definir la cantidad de iteraciones y el nombre del tablero
board= "simple"
i= 100
```

```python
df_gs= pd.read_excel(f'./results/{board}_greedy_search_{str(i)}_times.xlsx')
df_as= pd.read_excel(f'./results/{board}_A_star_{str(i)}_times.xlsx')
df_ids= pd.read_excel(f'./results/{board}_IDS_{str(i)}_times.xlsx')
df_dfs= pd.read_excel(f'./results/{board}_DFS_{str(i)}_times.xlsx')
df_bfs= pd.read_excel(f'./results/{board}_BFS_{str(i)}_times.xlsx')
```

```python
df= pd.concat([df_bfs, df_dfs, df_ids, df_gs, df_as])
```
